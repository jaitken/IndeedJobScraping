```python
import pandas as pd
from splinter import Browser #import packages to access and scrape web browsers
from selenium import webdriver #used to click through browser 
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
import pymongo
import datetime
#mongo connection string here
client = pymongo.MongoClient("")
db = client.Jobs
fullStack = db.fullStack
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
def get_Page_Number(link):   
    
    a_string = link.replace(',', '')
    numbers = []
    for word in a_string.split():
        
        if word.isdigit():
            
            numbers.append(int(word))
    return int(numbers[1]/50)+1 #get page number based off number of items showing
```

```python
def getInfo(descNum):

    with browser.get_iframe("vjs-container-iframe") as iframe: #goes into iframe in the html 
        
        #xpath of the text needed
        jobDesc = browser.find_by_xpath('/html/body').text
        splitJobDesc = jobDesc.splitlines()
        parsedDesc = jobDesc.replace('\n', ' ').replace('\r', '')
        
      
        document = {
          "title": splitJobDesc[0],
          "company": splitJobDesc[2],
          "description": parsedDesc
        }
        
        try:
    
            fullStack.insert_one(document)
            print(descNum, ':\nTitle: ', splitJobDesc[0], '\nCompany: ', splitJobDesc[2])
            print('-------------------------------------------------------------------------------------------------------\n')
        except Exception as e:
            
            print('Exception: ',e, '\nFailed to insert')
```

```python
def open_Info(): 
    
    
#-------------------------------------------SEARCH TERM IN QUOTES HERE--------------------------------------------------------#    
    b = browser.find_link_by_partial_text('Full Stack') #change depending on position you are searching
#-----------------------------------------------------------------------------------------------------------------------------#    
    
    
    #date = browser.find_by_css('span[class = "date date-a11y"]')
    
    count = 0
    for i in range(len(b)+1): #change depending on ?limit (number of items shown)
        
        time.sleep(15) #wait timer
        try:
            
            count += 1
            
#--------------------------------------------CHANGE THE NUMBER IF FIRST JOB POSTING IS NOT BEING PULLED------------------------#            
            b[i+2].click()#change this number based off headers that may give bad results
#------------------------------------------------------------------------------------------------------------------------------#             
            

            getInfo(count) #Call function that puts info into mongodb
        except Exception:
            count -= 1
            pass
```

```python
def start(pageNum):
    
    
#------------------------------------------------------CHANGE URL HERE---------------------------------------------------------#
    link = 'https://www.indeed.com/jobs?q=full%20stack&sort=date&limit=50&start=' #?start = page * limit, ?q= *search term here*
#------------------------------------------------------------------------------------------------------------------------------#    
        
    #for j in range(get_Page_Number(browser.find_by_id('searchCountPages').text)): 
    
    num = (pageNum-1)*50
    full = link+str(num)
    browser.visit(full) #visit website
    open_Info()
    print('Finished looking through page ', (pageNum-1))
    
    
```

```python
import time
from tkinter import * 
try:
    import Tkinter as tk
except:
    import tkinter as tk
    
class Test():
    
    def __init__(self):
        
        self.root = tk.Tk()
        self.root.geometry('300x200')
        
        label = Label(self.root, text = 'Enter Page Number')
        label.pack(pady = 5)
        
        def returnEntry(arg=None):

            result = myEntry.get()
            myEntry.delete(0,END)
            print('Page Number: ', result)
            start(int(result))

        # Create the Entry widget
        myEntry = Entry(self.root, width=20)
        myEntry.focus()
        myEntry.bind("<Return>",returnEntry)
        myEntry.pack(pady = 10)

        # Create the Enter button
        enterEntry = Button(self.root, text= "Enter", command=returnEntry)
        enterEntry.place(x=100, y = 50)
        enterEntry.pack(pady = 10)
        
        quitbutton = tk.Button(self.root, 
                          text = 'Quit', 
                          command= self.quit)
        quitbutton.place(x=100, y=100)
        
        quitbutton.pack(pady = 10)
        self.root.mainloop()
        
    def quit(self):
        self.root.destroy()
        
```
